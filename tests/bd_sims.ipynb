```python
import pickle as pkl

def LoadPKLData(trajOutName="test.pkl"):
    file = open(trajOutName,"rb")
    data = pkl.load(file)
    file.close()

    [ts,xs,ys] = data
    return ts,xs,ys

```

```python
ts,xs,ys = LoadPKLData(trajOutName)
nUpdates = np.shape(ts)[0]
nParticles = np.shape(xs)[0]

```

```python
# plot no more than three trajectories 
dParticles = np.min([3,nParticles])
for i in range(dParticles):
  plt.plot(xs[i,:],ys[i,:])

for i in range(dParticles):
  plt.scatter(xs[i,-1],ys[i,-1])
#plt.xlim([-1,2])
#plt.ylim([0,2])
```

```python
x0s = xs[:,0]
y0s = ys[:,1]
msds = np.zeros(nUpdates)
for i in range(nUpdates):
  #j = 0
  #xmsd = xs[j,i] - x0s[j]
  #ymsd = ys[j,i] - y0s[j]
  xmsd = xs[:,i] - x0s
  ymsd = ys[:,i] - y0s

  sd = xmsd*xmsd + ymsd*ymsd
  msd = np.mean(sd)
  msds[i] = msd
```
