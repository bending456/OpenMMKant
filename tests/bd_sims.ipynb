```python
nParticles=10
nUpdates = 100
#ts,xs,ys = bd.runBD(display=True,nParticles=nParticles,nUpdates=nUpdates,addForce=False) #True)
ts,xs,ys = bd.runBD(display=True,nParticles=nParticles,nUpdates=nUpdates,addForce=True)
```

```python
dParticles = np.min([3,nParticles])
for i in range(dParticles):
  plt.plot(xs[i,:],ys[i,:])

for i in range(dParticles):
  plt.scatter(xs[i,-1],ys[i,-1])
#plt.xlim([-1,2])
#plt.ylim([0,2])
```

```python
x0s = xs[:,0]
y0s = ys[:,1]
msds = np.zeros(nUpdates)
for i in range(nUpdates):
  #j = 0
  #xmsd = xs[j,i] - x0s[j]
  #ymsd = ys[j,i] - y0s[j]
  xmsd = xs[:,i] - x0s
  ymsd = ys[:,i] - y0s

  sd = xmsd*xmsd + ymsd*ymsd
  msd = np.mean(sd)
  msds[i] = msd
```

```python
plt.plot(msds)
```
